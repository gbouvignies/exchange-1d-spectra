## 1. The Model

In this section, we examine a protein undergoing chemical exchange between two states, **A** and **B**, each characterized by distinct chemical environments. The exchange process is represented by the equation:

$$
A \underset{k_\text{BA}}{\stackrel{k_\text{AB}}{\rightleftharpoons}} B \tag{1}
$$

Here, $k_\text{AB}$ and $k_\text{BA}$ are the forward and reverse first-order rate constants, respectively. The sum of these rate constants, $k_\text{ex} = k_\text{AB} + k_\text{BA}$, represents the overall exchange rate constant. The fractional populations of states **A** and **B** are denoted as $p_\text{A}$ and $p_\text{B}$, respectively, with the constraint that $p_\text{A} + p_\text{B} = 1$. The lifetimes of states **A** and **B** are defined as $\tau_A = 1/k_\text{AB}$ and $\tau_B = 1/k_\text{BA}$. The timescale of the exchange process is defined as $\tau_{ex} = 1 / k_\text{ex}$.

In the subsequent discussion, we focus on a specific nucleus within the protein. For this nucleus, the chemical shift (measured in rad·s<sup>-1</sup>) is $\omega_\text{A}$ in state **A** and $\omega_\text{B}$ in state **B**, with a difference of $\Delta \omega_\text{AB} = \omega_\text{B} - \omega_\text{A}$.

---


## 2. The Kinetic Equations

The system of differential equations for the first-order, two-site exchange reaction shown in Equation (1) can be formulated using the chemical reaction rate law:

$$
\frac{d}{dt}[A] = - k_\text{AB}[A] + k_\text{BA}[B] \tag{2}
$$

$$
\frac{d}{dt}[B] = - k_\text{BA}[B] + k_\text{AB}[A] \tag{3}
$$

These equations can be expressed in matrix form as:

$$
\frac{d}{dt} \begin{pmatrix} [A] \\ [B] \end{pmatrix} =
\begin{pmatrix}
    -k_\text{AB} & k_\text{BA} \\
    k_\text{AB} & -k_\text{BA} \\
\end{pmatrix}
\begin{pmatrix} [A] \\ [B] \end{pmatrix}
\tag{4}
$$

At equilibrium, the concentrations of A and B are constant, so:

$$
\frac{d}{dt}[A]=\frac{d}{dt}[B]=0 \implies \frac{{[A]}_\text{eq}}{{[B]}_\text{eq}}=\frac{k_\text{BA}}{k_\text{AB}} \tag{5}
$$

which leads to the fractional population of state B:

$$
p_\text{B} = \frac{{[B]}_\text{eq}}{{[A]}_\text{eq}+{[B]}_\text{eq}} = \frac{k_\text{AB}}{k_\text{AB}+k_\text{BA}} \tag{6}
$$

---


## 3. The Bloch Equations

The Bloch equations describe the evolution of magnetization in the rotating frame (with the static $B_0$ field aligned along the z-direction). These can be represented in matrix form as:

$$
\frac{d}{dt} \begin{pmatrix} M_\text{x} \\ M_\text{y} \\ M_\text{z} \end{pmatrix} =
\begin{pmatrix}
    -R_2 & -\Omega & \omega_1\sin\phi \\
    \Omega & -R_2 & -\omega_1\cos\phi \\
    -\omega_1\sin\phi & \omega_1\cos\phi & -R_1 \\
\end{pmatrix}
\begin{pmatrix} M_\text{x} \\ M_\text{y} \\ M_\text{z} \end{pmatrix} +
\begin{pmatrix} 0 \\ 0 \\ R_1 M_\text{eq} \end{pmatrix}
\tag{7}
$$

Here, $R_1$ and $R_2$ represent the rates at which the longitudinal ($M_z$) and transverse ($M_x, M_y$) magnetization components return to equilibrium (in s<sup>-1</sup>), respectively. $\Omega$ is the offset resonance frequency (in rad·s<sup>-1</sup> from the carrier frequency), and $\omega_1$ and $\phi$ are the strength (in rad·s<sup>-1</sup>) and phase of the applied RF field, respectively. $M_{i}$ denotes the magnetization component along the $i \in \{x,y,z\}$ axis, and $M_\text{eq}$ refers to the thermal equilibrium longitudinal magnetization (often denoted $M_0$).

To express the Bloch equations in a homogeneous form, suitable for matrix exponentiation, we augment the magnetization component vector with the identity operator $E$:

$$
\frac{d}{dt}
\begin{pmatrix} E \\ M_\text{x} \\ M_\text{y} \\ M_\text{z} \end{pmatrix} =
\begin{pmatrix}
    0 & 0 & 0 & 0 \\
    0 & -R_2 & -\Omega & \omega_1\sin\phi \\
    0 & \Omega & -R_2 & -\omega_1\cos\phi \\
    R_1M_\text{eq} & -\omega_1\sin\phi & \omega_1\cos\phi & -R_1 \\
\end{pmatrix}
\begin{pmatrix} E \\ M_\text{x} \\ M_\text{y} \\ M_\text{z} \end{pmatrix}
\tag{8}
$$

---


## 4. The Bloch-McConnell Equations

In a seminal 1958 article, McConnell recognized the analogy between the evolution of magnetization and chemical concentrations, demonstrating that the Bloch equations can be extended to include the effects of chemical exchange. The resulting generalized equations, known as the Bloch-McConnell equations, can be expressed in matrix form for a two-site exchange system as follows:

$$
\frac{d}{dt} \begin{pmatrix} E \\ M_\text{x}^\text{A} \\ M_\text{y}^\text{A} \\ M_\text{z}^\text{A} \\ M_\text{x}^\text{B} \\ M_\text{y}^\text{B} \\ M_\text{z}^\text{B} \end{pmatrix} =
\begin{pmatrix}
    0 & 0 & 0 & 0 & 0 & 0 & 0 \\
    0 & -R_{2}^\text{A} - k_\text{AB} & -\Omega^\text{A} & \omega_1\sin\phi & k_\text{BA} & 0 & 0 \\
    0 & \Omega^\text{A} & -R_{2}^\text{A} - k_\text{AB} & -\omega_1\cos\phi & 0 & k_\text{BA} & 0 \\
    R_{1}^\text{A}M_{eq}^\text{A} & -\omega_1\sin\phi & \omega_1\cos\phi & -R_{1}^\text{A} - k_\text{AB} & 0 & 0 & k_\text{BA} \\
    0 & k_\text{AB} & 0 & 0 & -R_{2}^\text{B} - k_\text{BA} & -\Omega^\text{B} & \omega_1\sin\phi \\
    0 & 0 & k_\text{AB} & 0 & \Omega^\text{B} & -R_{2}^\text{B} - k_\text{BA} & -\omega_1\cos\phi \\
    R_{1}^\text{B}M_{eq}^\text{B} & 0 & 0 & k_\text{AB} & -\omega_1\sin\phi & \omega_1\cos\phi & -R_{1}^\text{B} - k_\text{BA} \\
\end{pmatrix}
\begin{pmatrix} E \\ M_\text{x}^\text{A} \\ M_\text{y}^\text{A} \\ M_\text{z}^\text{A} \\ M_\text{x}^\text{B} \\ M_\text{y}^\text{B} \\ M_\text{z}^\text{B} \end{pmatrix} \tag{9}
$$

In Equation (9), state-specific parameters are denoted by the superscript $i \in \{ A, B \}$. For instance, $M_{eq}^\text{A} = p_\text{A} M_0$ and $M_{eq}^\text{B} = p_\text{B} M_0$, where $M_0$ is the total thermal equilibrium magnetization. This equation describes a single isolated spin-½ undergoing exchange between two chemical states. However, the same formalism can be extended to more complex spin systems and kinetic models.

The solution to a homogeneous first-order differential equation system of the form:

$$ \frac{d}{dt}\mathbf{M}(t) = \mathbf{L} \cdot \mathbf{M}(t) \tag{10} $$

is given by:

$$
\mathbf{M}(t+\Delta t) = \exp{\left(\mathbf{L}\Delta t \right)} \cdot \mathbf{M}(t) \tag{11}
$$

where $\mathbf{L}$ is the matrix of coefficients (often called the Liouvillian matrix in NMR) and $\mathbf{M}$ is the vector of (magnetization) components, as used in Equations (8) or (9). Equation (4) is an example of such a system for chemical kinetics.

To simulate the magnetization trajectory during an NMR experiment in the presence of chemical exchange, the propagator $\exp{\left( \mathbf{L} \Delta t \right)}$ needs to be computed for each individual part (e.g., delay, pulse) of the experiment and multiplied successively with the initial magnetization vector.

## 5. The 1D NMR Spectrum in Presence of Two-Site Exchange

To simulate the spectra typically obtained after a 90° pulse (assuming detection starts immediately after the pulse and relaxation during the pulse is negligible), we primarily need to consider the evolution of the transverse magnetization component, $M^+ = M_\text{x} + iM_\text{y}$. The Bloch-McConnell equations, when written in the laboratory frame for these components, can be simplified for a two-site system as:

$$
\frac{d}{dt} \begin{pmatrix} M_\text{A}^+ \\ M_\text{B}^+ \end{pmatrix} =
\begin{pmatrix}
    -R_{2}^\text{A} + i\omega_\text{A} - k_\text{AB} & k_\text{BA} \\
    k_\text{AB} & -R_{2}^\text{B} + i\omega_\text{B} - k_\text{BA} \\
\end{pmatrix}
\begin{pmatrix} M_\text{A}^+ \\ M_\text{B}^+ \end{pmatrix} \tag{12}
$$

The solution to this system is given by:

$$
\begin{pmatrix} M_\text{A}^+(t) \\ M_\text{B}^+(t) \end{pmatrix} =
\begin{pmatrix} a_{11}(t) & a_{12}(t) \\ a_{21}(t) & a_{22}(t) \end{pmatrix}
\begin{pmatrix} M_\text{A}^+(0) \\ M_\text{B}^+(0) \end{pmatrix} \tag{13}
$$

where the time-dependent matrix elements $a_{ij}(t)$ are given by:

$$
a_{11}(t) = \frac{1}{2}
\left[
    \left( 1 - \frac
        {i\Delta \omega - \Delta R_2 + k_\text{ex} (p_\text{B} - p_\text{A})}
        {\lambda_+ - \lambda_-}
    \right)
    e^{-\lambda_-t} +
    \left( 1 + \frac
        {i\Delta \omega - \Delta R_2 + k_\text{ex} (p_\text{B} - p_\text{A})}
        {\lambda_+ - \lambda_-}
    \right)
    e^{-\lambda_+t}
\right] \tag{14}
$$

$$
a_{22}(t) = \frac{1}{2}
\left[
    \left( 1 + \frac
        {i\Delta \omega - \Delta R_2 + k_\text{ex} (p_\text{B} - p_\text{A})}
        {\lambda_+ - \lambda_-}
    \right)
    e^{-\lambda_-t} +
    \left( 1 - \frac
        {i\Delta \omega - \Delta R_2 + k_\text{ex} (p_\text{B} - p_\text{A})}
        {\lambda_+ - \lambda_-}
    \right)
    e^{-\lambda_+t}
\right] \tag{15}
$$

$$
a_{12}(t) = \frac{k_\text{ex} p_\text{A}}{\lambda_+ - \lambda_-} \left( e^{-\lambda_-t} - e^{-λ_+t} \right) \tag{16}
$$

$$
a_{21}(t) = \frac{k_\text{ex} p_\text{B}}{\lambda_+ - \lambda_-} \left( e^{-\lambda_-t} - e^{-λ_+t} \right) \tag{17}
$$

Here, $\Delta R_2 = R_{2}^\text{B} - R_{2}^\text{A}$, and the eigenvalues $\lambda_±$ of the matrix in Equation (12) are given by:

$$
\lambda_± = \frac{1}{2} \left[
    -i (\omega_\text{A} + \omega_\text{B}) + R_{2}^\text{A} + R_{2}^\text{B} + k_\text{ex} \pm
    \sqrt{
        \left(
            i\Delta \omega - \Delta R_2 + k_\text{ex} (p_\text{B} - p_\text{A})
        \right)^2 +
        4 k_\text{ex}^2 p_\text{A} p_\text{B}
    }
\right] \tag{18}
$$

_Note: The real parts of $(-\lambda_\pm)$ correspond to the effective decay rates and their imaginary parts to the effective frequencies of the exchanging signals._

## 6. Exchange Regimes

<img src="Images/chemical_exchange.png" width="60%">

**Figure 1**: Spectra of a ¹⁵N nucleus undergoing chemical exchange in an 800 MHz magnet, simulated using the solutions to the Bloch-McConnell equations (e.g., Equation 13). Parameters: $\Delta \delta_\text{AB} = (\delta_\text{B} - \delta_\text{A}) = 3 \text{ ppm}$. Left panel: $p_\text{B} = 0.5$. Right panel: $p_\text{B} = 0.05$.

**Figure 1** shows a series of spectra calculated as a function of $k_\text{ex}$. The left panel illustrates the scenario where the populations of the two states are equal ($p_\text{A} = p_\text{B} = 0.5$). When the exchange is **slow** ($k_\text{ex} \ll |\Delta \omega_\text{AB}|$), two distinct peaks are present. These peaks broaden as the exchange rate increases and eventually merge into a single, very broad peak when $k_\text{ex} \approx |\Delta \omega_\text{AB}|$ (a phenomenon known as **coalescence**). In the **fast exchange** regime ($k_\text{ex} \gg |\Delta \omega_\text{AB}|$), a single, narrow peak appears at the population-weighted average chemical shift.

A more intricate situation arises when the populations are highly skewed, with $p_\text{A} \gg p_\text{B}$ (right panel of **Figure 1**). In the slow exchange regime, two peaks are present at positions $\omega_{A}$ and $\omega_{B}$, with their integrated intensities (areas) proportional to the populations $p_A$ and $p_B$. As $k_\text{ex}$ increases, the peak corresponding to the minor state (**B**) experiences significant line broadening and rapidly becomes difficult to detect. The peak corresponding to the major state (**A**), however, remains visible and also broadens as it approaches coalescence, albeit to a lesser extent than resonance **B**. As the exchange rate becomes even faster, this peak narrows and ultimately appears at the position corresponding to the weighted average of the two states.

The exchange regime is broadly determined by the ratio of the exchange rate ($k_\text{ex}$) to the chemical shift difference between the two states (in rad·s<sup>-1</sup>, $|\Delta \omega_\text{AB}|$). Thus, one can shift the system toward a slower (or faster) exchange regime by:

- Decreasing (or increasing) $k_\text{ex}$ (e.g., by changing temperature, pH, or ionic strength).
- Increasing (or decreasing) $|\Delta \omega_\text{AB}|$, which is directly proportional to the static magnetic field strength ($B_0$). Thus, experiments at different $B_0$ fields can help characterize exchange phenomena.

These observations can be further quantitatively described by examining the approximate solutions of the Bloch-McConnell equations for the two limiting cases.

### 6.1. Slow Exchange

In the slow exchange regime, where $k_\text{ex} \ll \left| i\Delta \omega_\text{AB} - \Delta R_2 + k_{ex}(p_B - p_A) \right|$ (more simply, $k_{ex} \ll |\Delta\omega_{AB}|$, especially if $\Delta R_2$ is small), Equation (13) simplifies to approximately give two independently decaying signals:

$$
M^+(t) \approx M_\text{A}^+(0) \exp\left[\left( i\omega_\text{A} - (R_2^\text{A} + k_\text{AB}) \right)t\right] +
M_\text{B}^+(0) \exp\left[\left( i\omega_\text{B} - (R_2^\text{B} + k_\text{BA}) \right)t\right] \tag{19}
$$

As expected, we observe two resonances at positions corresponding to states **A** and **B**. However, the exchange process contributes to the effective transverse relaxation rates of the two peaks, leading to larger linewidths ($\Delta\nu_{1/2} = R_{2,eff}/\pi$). The effective relaxation rates are $R_{\text{2,eff}}^\text{A} = R_2^\text{A} + k_\text{AB}$ and $R_{\text{2,eff}}^\text{B} = R_2^\text{B} + k_\text{BA}$ for states **A** and **B**, respectively.
In the case where $p_\text{A} \gg p_\text{B}$ (meaning $k_\text{BA} \gg k_\text{AB}$ since $p_B/p_A = k_{AB}/k_{BA}$), the exchange contribution to the linewidth of peak **B** ($k_\text{BA}$) becomes much larger than that for peak **A** ($k_\text{AB}$). For example, assuming $k_\text{ex} = 1000 \text{ s}^{-1}$ and $p_\text{B} = 0.02$ (i.e., 2%), we have $k_\text{AB} = p_B k_{ex} = 20 \text{ s}^{-1}$ and $k_\text{BA} = p_A k_{ex} = 980 \text{ s}^{-1}$. This makes peak **B** significantly broader and potentially undetectable in NMR spectra. Note that in this slow exchange limit, the exchange contribution to the linewidth ($k_{AB}$ or $k_{BA}$) is independent of the $B_0$ field (assuming $k_{ex}$ itself is field-independent).

### 6.2. Fast Exchange

In the fast exchange regime, where $k_\text{ex} \gg \left| i\Delta \omega_\text{AB} - \Delta R_2 + k_{ex}(p_B - p_A) \right|$ (more simply, $k_{ex} \gg |\Delta\omega_{AB}|$), Equation (13) approximates to a single decaying signal:

$$
M^+(t) \approx M^+(0) \exp\left[\left( i\overline{\omega} - \overline{R_2} -
\frac{p_\text{A}p_\text{B} (\Delta \omega_\text{AB})^2}{k_\text{ex}} \right)t\right] \tag{20}
$$

where $\overline{R_2} = p_\text{A} R_2^\text{A} + p_\text{B} R_2^\text{B}$ and $\overline{\omega} = p_\text{A} \omega_\text{A} + p_\text{B} \omega_\text{B}$ are the population-averaged transverse relaxation rate and chemical shift, respectively. Here, only a single line at the weighted-average position ($\overline{\omega}$) is observed. Its effective relaxation rate is the weighted average of the intrinsic rates ($\overline{R_2}$) plus an additional **exchange contribution to line broadening**, $R_{ex} = \frac{p_\text{A}p_\text{B} (\Delta \omega_\text{AB})^2}{k_\text{ex}}$.
Interestingly, while the exchange-induced line broadening in the slow exchange regime ($k_{AB}, k_{BA}$) is independent of the $B_0$ field, the $R_{ex}$ term in the fast exchange regime varies quadratically with $\Delta \omega_\text{AB}$ (since $\Delta\omega_{AB}$ is proportional to $B_0$), and thus as $B_0^2$. This difference in field dependence is a key characteristic used to distinguish between exchange regimes.

## 8. Kinetics of a Binding Equilibrium: E + L $\rightleftharpoons$ EL

Describing the effect of ligand binding on the NMR line shapes of a target molecule is a specific application of the two-state exchange model. The observed nucleus on the target molecule (E) can exist in either the free state (A) or the ligand-bound state (B). The relative populations of these states are defined by the dissociation equilibrium constant, $K_\text{D}$, and the total concentrations of the ligand, $L_0$, and the target molecule, $E_0$, according to the law of mass action:

$$
K_\text{D} = \frac{k_\text{off}}{k_\text{on}} = \frac{[E][L]}{[EL]} \tag{21}
$$

Here, $[E]$ and $[L]$ represent the equilibrium concentrations of the free target molecule and the free ligand, respectively.

If we denote $p_\text{B}$ as the bound fraction of E (i.e., $p_\text{B} = [EL]/E_0$), $p_\text{B}$ is a root of the following quadratic equation, which can be derived from Equation (21):

$$
p_\text{B}^2 - p_\text{B}\left(1 + \frac{L_0}{E_0} + \frac{K_\text{D}}{E_0}\right) + \frac{L_0}{E_0} = 0
\tag{22}
$$

In this binding model, the forward rate constant for a nucleus in E going from the free to the bound state is pseudo-first-order ($k_{AB} = k_\text{on}[L]$), while the reverse rate is first-order ($k_{BA} = k_\text{off}$). Therefore, the overall exchange rate constant, $k_\text{ex}$, is defined as:

$$
k_\text{ex} = k_\text{on}[L] + k_\text{off} = k_\text{on}\left([L] + K_\text{D}\right) \tag{23}
$$

The line shapes observed in a titration experiment depend on both the populations of the states (determined by $K_D$, $E_0$, and $L_0$) and the exchange rate constant $k_\text{ex}$ (determined by kinetic rates and the free ligand concentration $[L]$). Depending on the interplay of these parameters, three limiting exchange regimes are typically observed:

### 8.1. Slow Exchange

This situation is often encountered in systems with strong binding (i.e., a very small $K_\text{D}$), where the dissociation rate constant, $k_\text{off} = k_\text{on} \cdot K_D$, is consequently very small. If $k_\text{ex} = k_\text{on}[L] + k_\text{off}$ remains much smaller than the chemical shift difference ($k_\text{ex} \ll |\Delta\omega|$), the system is in slow exchange. During the NMR measurement, molecules are effectively "frozen" in either their bound or free states. As the ligand is titrated, two corresponding resonance lines are visible, with their relative intensities changing to reflect the populations of the free and bound species.

### 8.2. Fast Exchange

This regime is typical for weak to intermediate binding interactions where the kinetic rates, particularly the dissociation rate $k_\text{off}$, are high. When the overall exchange rate is much greater than the chemical shift difference ($k_\text{ex} \gg |\Delta\omega|$), the system is in fast exchange. This leads to a single observed resonance whose chemical shift ($\nu_\text{obs}$) is the population-weighted average of the free ($\nu_\text{free}$) and bound ($\nu_\text{bound}$) states. Thus, by measuring the chemical shift perturbation as a function of the total ligand concentration $L_0$, one can determine the bound fraction $p_\text{B}$:

$$
p_\text{B}(L_0) = \frac{\nu_\text{obs} - \nu_\text{free}}{\nu_\text{bound} - \nu_\text{free}} \tag{24}
$$

Fitting the experimental values of $p_\text{B}$ as a function of $L_0$ to Equation (22) provides a powerful way to determine the dissociation equilibrium constant, $K_D$. Note that while $\nu_\text{free}$ is easily determined from a sample without ligand, $\nu_\text{bound}$ may be challenging to measure directly, since reaching 100% saturation may not be feasible due to weak binding or solubility limits. Therefore, the non-linear fitting procedure often treats both $K_D$ and $\nu_\text{bound}$ as adjustable parameters.

### 8.3. Intermediate Exchange

This situation occurs when the overall exchange rate, $k_\text{ex}$, is on the same timescale as the chemical shift difference, $|\Delta\omega|$ (i.e., $k_\text{ex} \approx |\Delta\omega|$). This results in severe line broadening, where resonances can become so broad that they effectively vanish into the baseline upon addition of the ligand. In such cases, one can attempt to move the system out of this regime and into a faster or slower exchange regime by:

- Changing the temperature to alter the kinetic rates ($k_{on}, k_{off}$).
- Changing the static magnetic field strength, $B_0$, to alter the chemical shift difference $|\Delta \omega|$.
